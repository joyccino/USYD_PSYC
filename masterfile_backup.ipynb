```python
from google.colab import auth
import gspread
from google.auth import default
import math
import random
import pandas as pd
#autenticating to google
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
def quiz_init() :
  #defining my worksheet
  subject_worksheet = input("1 for PSYC 2016, 2 for PSYC 2017 and 3 for PSYC 3014")

  if subject_worksheet == "1" :
    subject_worksheet = gc.open('Final_prep').worksheet('PSYC2016')
  elif subject_worksheet == "2" :
    subject_worksheet = gc.open("Final_prep").worksheet('PSYC2017')
  else :
    subject_worksheet = gc.open("Final_prep").worksheet('PSYC3014')

  #get_all_values gives a list of rows
  rows = subject_worksheet.get_all_values()
  #Convert to a DataFrame 
  df = pd.DataFrame(rows)
  df = df.rename(columns=df.iloc[0]).drop(df.index[0])

  # 문제 실행
  i = 0
  corrects = 0

  
    
  # 문제는 몇 퍼센트?
  no_of_questions = round(len(df)*int(input("percent of the questions."))/100)
  
  print()
  if (no_of_questions < 1) :
    no_of_questions = 1
  print()
  print("We have ",no_of_questions," questions for you. Let's begin.")

  # 문제 옵션 섞기
  df_shuffled = df.sample(frac=1).reset_index(drop=True)

  df_shuffled = df_shuffled.iloc[:no_of_questions]
    
  while(i < len(df_shuffled)) :
    #문제 출력
    print("question no. ",i+1,") ",df_shuffled['question'][i])
    options = [df_shuffled['opt1'][i], df_shuffled['opt2'][i], df_shuffled['opt3'][i], df_shuffled['opt4'][i],df_shuffled['opt5'][i]]
    random.shuffle(options)

    while("" in options):
      options.remove("")
  
    print()

    for opt in options:
      print(opt)

    print()
    user_input = input("Enter your question:")
    

    if user_input == df_shuffled['answer'][i] :
      print("")
      print("correct!")
      # find cell
      prev_rate = int(df_shuffled['correct'][i])
      q_no = "J"+df_shuffled['index'][i]
      subject_worksheet.update(q_no, prev_rate+1)


      corrects += 1
    else :
      print("")
      print("wrong, correct answer is => ",df_shuffled['answer'][i])
      # find cell
      prev_rate = int(df_shuffled['missed'][i])
      q_no = "K"+df_shuffled['index'][i]
      subject_worksheet.update(q_no, prev_rate+1)

    print("="*50)

    i += 1


  # no of questions you got right : wrongs
  print("corrects : no of questions")
  print(corrects, " : ", len(df_shuffled) - corrects)
  print("a percentage of correct answers: ",corrects/len(df_shuffled) *100, "%")       

```
