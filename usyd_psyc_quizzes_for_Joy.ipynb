```python
#@title PSYC2017 - SAQ session.

# short answer questions.

import spacy
nlp = spacy.load("en_core_web_lg")

#@title Short answer questions
from google.colab import auth
import gspread
from google.auth import default
import math
import random
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

#subject_worksheet = gc.open('Final_prep_Joy_11_Nov').worksheet('PSYC2016')
df = pd.read_excel('/content/drive/MyDrive/PSYC2017_prep.xlsx')

df = df.fillna(0)

df_original = df

# 문제 실행
i = 0
corrects = 0

# 문제는 몇개?
no_of_questions = input("number of questions? \n")
print("\nWe have ",no_of_questions," questions for you. Let's begin.\n")

df['index'] = np.arange(0, len(df))
df_original['index'] = np.arange(0, len(df))

# 문제 옵션 섞기
df_shuffled = df.sample(frac=1).reset_index(drop=True)
df_shuffled = df_shuffled.iloc[:int(no_of_questions)]
  
while(i < len(df_shuffled)) :
  #문제 출력
  print(df_shuffled['lecturer'][i],":",df_shuffled['type'][i]," ",df_shuffled['topic'][i])

  print("[Question no. {0}]".format(i+1),df_shuffled['question'][i])
  print('-'*50)
  
  user_input = input("Enter your answer:\n")

  confident1 = input("are you confident? y/n \n")

  answer = df_shuffled['answer'][i]

  print(answer+"\n")
  calc = nlp(answer).similarity(nlp(user_input))
  print("Answer Similarity = ",calc)

  confident2 = input("\nAre you still confident? y/n \n")

  

  if(confident1 == "y" and confident2 == "y") :
    prev_confident = int(df_shuffled['confident'][i])
    df_original.at[int(df_shuffled['index'][i]), 'confident'] = prev_confident + 1
    
    corrects += 1

  else :
   
    prev_rate = int(df_shuffled['notconfident'][i])
    df_original.at[int(df_shuffled['index'][i]), 'notconfident'] = prev_rate + 1

  print("="*50)

  i += 1

# no of questions you got right : wrongs
print("\ncorrects : missed")
print(corrects, " : ", len(df_shuffled) - corrects)
print("\na percentage of correct answers: ",corrects/len(df_shuffled) *100, "%")  

df_original = df_original.drop(columns = "index")
df_original.to_excel('/content/drive/MyDrive/PSYC3014_prep.xlsx', index=False)

unanswered = df_original[(df_original['confident'] == 0) & (df_original['notconfident']==0)]
print("You have {0} ({1}%) unanswered questions.".format(len(unanswered),len(unanswered)/len(df_original)*100))

if(len(unanswered) > 0) :
  print("Keep it up!")
```
